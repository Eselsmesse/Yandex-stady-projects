# Описание проекта

*Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). На основе которых нужно понять тенденции влияющие на общие продажи игр и спланировать рекламные компании и вектор разработки на 2017 год.*

**Цель:** 

Проанализировать данные влияющие на успешность игр и на основе этого анализа спланировать рекламные кампании на 2017 год. 

**Ход исследования**

Данные о выпущеных играх содержатся в файле `games.csv`. В данных имеется всё что нам нужно для анализа, но имеется большое количество пропусков которые нужно будет обработать и создать выборки на основе которых можно будет строить прогнозы.
 
Таким образом, исследование пройдёт в три этапа:
 1. Подготовка данных.
 2. Проведение исследовательских работ.
 3. Составление портрета пользователя
 4. Проверка гипотез.

Описание данных:

* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.
Данные за 2016 год могут быть неполными.

Сайт метакритика даёт такой ответ:

WHAT DOES "TBD" MEAN?

To Be Determined (будет определено).

To help ensure that METASCORES accurately reflect the reviews given by critics for any particular movie, game, television show or album, we do not display a METASCORE for those items that do not have at least four (4) reviews in our database. Once this minimum number of reviews is reached, the METASCORE will display.

*Исходя из этой информации принимаем значения 'tbd' как пустые (np.NaN), т.к. восстановить эти данные невозможно.*

## Обработка user_score и critic_score

```python
# заменяем значение tbd
game_data['user_score'] = game_data['user_score'].replace('tbd', np.NaN)
game_data['user_score'] = game_data['user_score'].astype('float', errors='ignore')
# умножаем user_score на 10 чтобы удобнее соотносить его с рейтингом критиков
game_data['user_score'] = game_data['user_score'] * 10 
game_data['user_score'] = game_data['user_score'].astype(pd.Int64Dtype()) # итоговое преобразование колонки
```

```python
# меняем тип данных критиков
game_data['critic_score'] = game_data['critic_score'].astype(pd.Int64Dtype())
```

**Вывод:**  
*Пропуски в данных обусловлены падением интереса к устаревшим платформам и играм на их базе.*  
*Чтобы не потерять такой большой объем данных во всех колонках с пропусками рейтингов (`critic_score`,`user_score`,`rating`), учитывая невозможность его восстановления по имеющимся данным, а так же за неимением возможности связаться с заказчиком, для уточнения, оставим их заполненными специальным значением -1(так как в колонке `user_score` есть нулевые значения, не будем их смешивать) и будем учитывать это в ходе работы и тоже самое проделаем с `year_of_release`. Это поможет нам сохранить данные и не повлиять на результаты исследования.*

```python
def miss_persent(data):
    """Функция считает процент пропущенных значений
    от общего количества
    """
    for col in data.columns:
        pct_missing = np.mean(data[col].isna())
        print('{} - {}%'.format(col, round(pct_missing*100,2)))
```

```python
miss_persent(game_data)
```

```python
# удаляем все пропущенные значения в year_of_release и name
# заполняем все пропуски в данных year_of_release, critic_score и user_score специальным значением -1
# так как колонка rating имеет тип object то пропуски в ней заполняем значением empty
game_data = game_data.dropna(subset=['name'])
game_data = game_data.dropna(subset=['year_of_release'])
game_data['year_of_release'] = game_data['year_of_release'].astype(int)
game_data['critic_score'] = game_data['critic_score'].fillna(-1)
game_data['user_score'] = game_data['user_score'].fillna(-1)
game_data['rating'] = game_data['rating'].fillna('empty')
```

```python
# проверка
miss_persent(game_data)
```

## Посмотрите, как менялись продажи по платформам. Выберите платформы снаибольшими суммарными продажами и постройте распределение по годам.


```python
platform_game_sales = game_data.pivot_table(index=['platform'], 
                                           columns='year_of_release', 
                                           values='all_sales', 
                                           aggfunc='sum', 
                                           fill_value='-')
```

```python
platform_game_sales
```

*По таблице прекрасно видно как одна технология приходит на рынок, достигает своего пика и планомерно изживает себя освобождаю дорогу для нового продукта*

```python
overall_platform_sales = game_data.pivot_table(index=['platform'],
                      values='all_sales',
                      aggfunc='sum').sort_values(by='all_sales', ascending=False)
overall_platform_sales
```

*В таблице видно существенный разрыв в количестве продажах между 6 и 7 местом, продажи PS в 2 раза больше чем у PS4. По симу построим графики продаж по первым 6-ти платформам.*

```python
# возбмём топ платформы и приближенные к нашему времени года в угоду актуальности подсчёта
top_platform_list = overall_platform_sales.index[0: 7]
actual_years = list(range(2012, 2017))
```

```python
overall_platform_sales_pivot = game_data.query('platform in @top_platform_list and year_of_release in @actual_years').pivot_table(index=['platform', 'year_of_release'], 
                                                              values='all_sales', aggfunc='sum',
                                                              fill_value='-')
overall_platform_sales_pivot
```

```python
overall_platform_sales_pivot.plot(y='all_sales', kind='bar', figsize=(15, 10))

plt.xlabel("Консоль, год", fontsize=15)
plt.ylabel("Суммарные продажи, млн.", fontsize=15)
plt.title("Продажи и время существование топ консолей", fontsize=20)
plt.show()
```

### Вывод

*Как мы можем заметить на рынке главенствуют PS и XBOX, так продолжается и по сей день, но продажи в PS суммарно идут лучше чем в xbox. Относительно этого тренда мы можем сказать что сейчас, в среднем, до того как продажи значительно упадут, срок жизни консоли составляет 3-4 года.*

```python
# функция позволяет отображать две таблицы
def display_side_by_side(*args):
    """Функция позволяет выводить несколько таблиц в ряд
    """
    html_str=' '
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

# Проверка гипотез

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.

Поясните:
* Как вы сформулировали нулевую и альтернативную гипотезы;
* Какой критерий применили для проверки гипотез и почему.

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

*Нам интересна двусторонняя гипотеза, потому что в условиях не сказано больше или меньше должен быть тот или иной рейтинг*

**Нулевая гипотеза H0** - Средний пользовательский рейтинг Xbox One и PC одинаковы.

**Альтернативная гипотеза H1** - Средние пользовательские рейтинги Xbox One и PC не равны.

```python
# определяем выборки для проверки гипотезы
PC_select = current_period.query('platform == "PC" and user_score != -1')
Xone_select = current_period.query('platform == "XOne" and user_score != -1')
```

```python
alpha = 0.05
# задаём критический уровень теоритической значимости, 
# если p-value будет превышать это значения значит теорию можно будет отвергнуть
```

```python
results_test = st.ttest_ind(Xone_select['user_score'],
                         PC_select['user_score'], equal_var=False
)

print('p-значение: ', results_test.pvalue)

if results_test.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

*Нам интересна двусторонняя гипотеза, потому что в условиях не сказано больше или меньше должен быть тот или иной рейтинг*

**Нулевая гипотеза H0** - Средние пользовательские рейтинги жанра Action и Sports одинаковые.

**Альтернативная гипотеза H1** - Средний пользовательский рейтинг Action и Sports разные.

```python
# определяем выборки для проверки гипотезы
Action_select = current_period.query('genre == "Action" and user_score != -1')
Sports_select = current_period.query('genre == "Sports" and user_score != -1')
```

```python
alpha = 0.05
# задаём критический уровень теоритической значимости, 
# если p-value будет превышать это значения значит теорию можно будет отвергнуть
```

```python
results_test = st.ttest_ind(Action_select['user_score'],
                         Sports_select['user_score'], equal_var=False
)

print('p-значение: ', results_test.pvalue)

if results_test.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод

Мы проверили 2 гипотезы:
1. Гипотеза о том что средние пользовательские рейтинги платформ Xbox One и PC одинаковые - **не отвергнута**.
2. Гипотеза о том что средние пользовательские рейтинги жанров Action и Sports разные - **отвергнута**
